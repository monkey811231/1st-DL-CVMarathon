### 1. 取出知乎問題發問時間

```python
pip install bs4
```

```python
# 1. 取出知乎問題發問時間
import requests
from bs4 import BeautifulSoup
import json

url = 'https://www.zhihu.com/api/v4/questions/55493026/answers'
headers = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.111 Safari/537.36'}
req = requests.get(url,headers=headers)
soup = BeautifulSoup(req.text,'html.parser')
req.text

json.loads(req.text)

```

```python
import json 
import datetime 
json_content = json.loads(req.text)
for i in json_content['data']:
    created = int(i['question']["created"])
    created_time = datetime.datetime.fromtimestamp(created)
    print(f'發問時間:{created_time}')
```

### 2. 取出第一筆與最後一筆回答的時間

```python
# 2. 取出第一筆與最後一筆回答的時間
import json 
import datetime 
json_content = json.loads(req.text)
print(type(json_content))
last_answer = datetime.datetime.fromtimestamp(int(json_content['data'][0]["updated_time"]))
frist_answer = datetime.datetime.fromtimestamp(int(json_content['data'][-1]["updated_time"]))

print(f'第一筆:{frist_answer} 最後一筆:{last_answer}')
```
