```python
# 1. 請問高雄市有多少地區有溫度資料？
with open('./data/64_72hr_CH.xml','r',encoding="utf-8") as f :
    print(f.read())  #用IO看

    doc = xmltodict.parse(f.read())
    print('==============')
    locations = doc['cwbopendata']['dataset']['locations']['location']
    print('高雄共有{}個地區有溫度資料'.format(len(locations)))
    
    for i in locations :
        print(i['locationName'])
 
```

```python
# 2. 請取出每一個地區所記錄的第一個時間點跟溫度
with open('./data/64_72hr_CH.xml','r',encoding="utf-8") as f :
#     print(f.read())
    doc = xmltodict.parse(f.read())
#     print(doc)
    locations = doc['cwbopendata']['dataset']['locations']['location']
    for i in locations :
        locationname = i['locationName']
#         print(i)
        locationdatatime = i['weatherElement'][0]['time'][0]['dataTime'].split('+')[0]
        locationTemp = i['weatherElement'][0]['time'][1]['elementValue']['value']
        
        print(f'地區 : {locationname}   \n\t第一紀錄時間/溫度 : {locationdatatime}  /  {locationTemp}度C\n')
    
```

```python
# 3. 請取出第一個地區所記錄的每一個時間點跟溫度
with open('./data/64_72hr_CH.xml','r',encoding="utf-8") as f :
#     print(f.read())
    doc = xmltodict.parse(f.read())
#     print(doc)
    locations = doc['cwbopendata']['dataset']['locations']['location'][0]
    print(locations['locationName'])
    for i in locations['weatherElement']:
        
        for j in i['time'] :
#             print(j)
            try:
                time = j['dataTime'].split('+')[0]
                temp = j['elementValue']['value']
                
                print(f'日期時間 : {time} \n\t 攝氏溫度 : {temp}度')
            except:
                pass
```
